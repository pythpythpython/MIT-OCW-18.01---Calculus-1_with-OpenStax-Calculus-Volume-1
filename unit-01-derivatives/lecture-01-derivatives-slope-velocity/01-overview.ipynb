## Overview

This lecture introduces the fundamental concept of the derivative, one of the two central ideas in calculus (along with the integral). We explore how derivatives arise naturally from the geometric problem of finding tangent lines and the physical problem of calculating instantaneous velocity.

### Learning Objectives

By the end of this lecture, you should be able to:

1. **Recognize the meaning of the tangent to a curve at a point**
   - Understand the geometric interpretation of a tangent line
   - Distinguish between secant and tangent lines
   - Visualize the limiting process as secant lines approach the tangent

2. **Calculate the slope of a tangent line**
   - Use the limit definition to find slopes
   - Apply both forms of the difference quotient
   - Compute tangent line equations

3. **Identify the derivative as the limit of a difference quotient**
   - Understand the two equivalent definitions
   - Recognize the notation $f'(a)$, $\frac{dy}{dx}$, $\frac{df}{dx}$
   - Apply the limit definition to specific functions

4. **Calculate the derivative of a given function at a point**
   - Use Equation 3.5: $f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$
   - Use Equation 3.6: $f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}$
   - Estimate derivatives numerically from tables

5. **Describe velocity as a rate of change**
   - Connect position, velocity, and derivatives
   - Understand the physical interpretation of instantaneous rate of change
   - Apply derivatives to motion problems

6. **Explain the difference between average and instantaneous velocity**
   - Compute average velocity over time intervals
   - Find instantaneous velocity as a limit
   - Relate this to the derivative of position

7. **Estimate derivatives from tables of values**
   - Use numerical approximation techniques
   - Understand the relationship between accuracy and interval size
   - Apply to real-world data

## Historical Context

Calculus was developed independently by **Isaac Newton (1643-1727)** and **Gottfried Leibniz (1646-1716)** in the late 17th century. While both mathematicians deserve credit for their groundbreaking work, their approaches differed:

- **Newton** focused on physics and motion, developing calculus to understand planetary motion and gravity
- **Leibniz** approached calculus more geometrically and developed much of the notation we use today

Both built upon earlier work by:
- **Pierre de Fermat** (method of finding maxima and minima)
- **Isaac Barrow** (geometric methods for tangents)
- **Bonaventura Cavalieri** (method of indivisibles)

The key insight of Newton and Leibniz was recognizing the **fundamental relationship between differentiation and integration** - that they are inverse operations.

## Key Concepts Preview

### The Big Idea

The derivative measures **instantaneous rate of change**. Just as:

- The slope of a line measures how steeply it rises or falls
- The velocity of an object measures how fast its position changes

The derivative $f'(a)$ measures how rapidly the function $f(x)$ changes at the point $x = a$.

### The Two Perspectives

**Geometric:** The derivative is the slope of the tangent line to the curve $y = f(x)$ at a point.

**Physical:** The derivative is the instantaneous velocity (rate of change of position with respect to time).

These two perspectives are mathematically equivalent, united by the concept of **limit**.
