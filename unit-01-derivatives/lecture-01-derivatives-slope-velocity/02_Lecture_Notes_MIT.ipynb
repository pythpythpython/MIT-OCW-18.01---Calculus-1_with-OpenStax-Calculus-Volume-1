## 1. The Tangent Line Problem

**Central Question**: How do we find the slope of a tangent line to a curve at a point?

### Approach: Secant Line Approximation

Given $f(x)$ and a point $(x_0, f(x_0))$:

1. Choose a nearby point $(x_0 + \Delta x, f(x_0 + \Delta x))$
2. Draw the **secant line** through both points
3. The slope of this secant is:

$$m_{sec} = \frac{f(x_0 + \Delta x) - f(x_0)}{\Delta x} = \frac{\Delta f}{\Delta x}$$

4. As $\Delta x \to 0$, the secant approaches the **tangent line**

## 2. Definition of the Derivative

### Formal Definition

The **derivative** of $f$ at $a$, denoted $f'(a)$, is:

$$f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$$

**Alternative form**:

$$f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}$$

provided these limits exist.
