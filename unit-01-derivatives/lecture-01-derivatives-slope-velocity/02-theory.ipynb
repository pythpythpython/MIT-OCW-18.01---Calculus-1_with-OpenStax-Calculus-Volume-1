## 1. Tangent Lines and Secant Lines

### Secant Lines

A **secant line** to a function $f$ is a line that intersects the graph of $f$ at two distinct points.

Given two points $(a, f(a))$ and $(x, f(x))$ on the graph of $f$, the slope of the secant line through these points is:

$$m_{\text{sec}} = \frac{f(x) - f(a)}{x - a}$$

This is called the **average rate of change** of $f$ over the interval $[a, x]$.

### Tangent Lines

Intuitively, a **tangent line** to a curve at a point $P$ is the line that "just touches" the curve at $P$ and has the same direction as the curve at that point.

**Key Insight:** As the second point moves closer to the first along the curve, the secant lines approach the tangent line.

### The Limiting Process

The slope of the tangent line at $(a, f(a))$ is obtained by taking the limit of the slopes of secant lines as the second point approaches $(a, f(a))$:

$$m_{\text{tan}} = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$$

provided this limit exists.

## 2. The Difference Quotient

### Definition

> **Definition 3.1 (Difference Quotient)**
>
> Let $f$ be a function defined on an interval $I$ containing $a$. 
>
> 1. If $x \neq a$ is in $I$, then
>    $$Q = \frac{f(x) - f(a)}{x - a}$$
>    is a **difference quotient**.
>
> 2. If $h \neq 0$ is chosen so that $a + h$ is in $I$, then
>    $$Q = \frac{f(a+h) - f(a)}{h}$$
>    is a **difference quotient with increment** $h$.

### Two Equivalent Forms

Both expressions measure the same thing - the average rate of change. The choice of which to use depends on convenience:

- **Form 1** (using $x$): $\frac{f(x) - f(a)}{x - a}$
  - Useful when $x$ is specified directly
  - Natural for geometric interpretations

- **Form 2** (using $h$): $\frac{f(a+h) - f(a)}{h}$
  - Useful for algebraic manipulations
  - $h$ represents the "change" in $x$
  - As $h \to 0$, we approach the point $a$

**Relationship:** Setting $x = a + h$ (or equivalently $h = x - a$) converts one form to the other.

## 3. Definition of the Derivative at a Point

### Formal Definition

> **Definition 3.2 (The Derivative at a Point)**
>
> Let $f$ be a function defined in an open interval containing $a$. The **derivative** of $f$ at $a$, denoted $f'(a)$, is defined by
>
> $$f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a} \tag{3.5}$$
>
> provided this limit exists.

If the limit exists, we say that $f$ is **differentiable** at $a$.

### Geometric Interpretation

The derivative $f'(a)$ equals the **slope of the tangent line** to the graph of $y = f(x)$ at the point $(a, f(a))$.

### Physical Interpretation

If $s(t)$ represents position at time $t$, then $s'(a)$ represents the **instantaneous velocity** at time $t = a$.

### The Process of Differentiation

Finding a derivative is called **differentiation**. To find $f'(a)$ using the definition:

1. Write out the difference quotient: $\frac{f(x) - f(a)}{x - a}$
2. Simplify the numerator algebraically
3. Cancel the factor $(x-a)$ from numerator and denominator
4. Take the limit as $x \to a$

## 4. Alternative Forms of the Derivative

### The $h$-form

> **Definition 3.3 (Alternative Definition)**
>
> Alternatively, we may define the derivative of $f$ at $a$ as
>
> $$f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h} \tag{3.6}$$
>
> provided this limit exists.

### Notation for Derivatives

The derivative can be expressed using various notations, each with its own advantages:

1. **Lagrange's notation**: $f'(a)$ or $f'(x)$
   - Simple and concise
   - Clear which function is being differentiated

2. **Leibniz's notation**: $\frac{df}{dx}$ or $\frac{dy}{dx}$
   - Emphasizes the ratio of infinitesimal changes
   - Useful for manipulating differentials
   - Common in physics and engineering

3. **Newton's notation**: $\dot{f}$ (for time derivatives)
   - Used primarily in physics
   - Indicates differentiation with respect to time

4. **Operator notation**: $\frac{d}{dx}f(x)$ or $D_x f(x)$
   - Treats differentiation as an operation

### At a specific point

To evaluate the derivative at $x = a$:

$$\left. \frac{dy}{dx} \right|_{x=a} \quad \text{or} \quad f'(a)$$

## 5. Velocity and Rates of Change

### Position, Velocity, and Acceleration

If $s(t)$ represents the position of an object at time $t$:

- **Average velocity** over $[a, t]$:
  $$v_{\text{ave}} = \frac{s(t) - s(a)}{t - a} \tag{3.7}$$

- **Instantaneous velocity** at $t = a$:
  $$v(a) = s'(a) = \lim_{t \to a} \frac{s(t) - s(a)}{t - a} \tag{3.8}$$

- **Speed** at $t = a$:
  $$\text{speed} = |v(a)|$$

### General Rates of Change

> **Definition 3.4 (Instantaneous Rate of Change)**
>
> The **instantaneous rate of change** of a function $f(x)$ at a value $a$ is its derivative $f'(a)$.

The derivative measures how rapidly $f$ changes at $x = a$. Applications include:

- **Physics**: velocity, acceleration, force
- **Economics**: marginal cost, marginal revenue, marginal profit
- **Biology**: population growth rate, reaction rate
- **Chemistry**: rate of reaction
- **Engineering**: heat flow, current

## 6. Differentiability and Continuity

### Theorem: Differentiability Implies Continuity

> **Theorem 3.1**
>
> Let $f(x)$ be a function and $a$ be in its domain. If $f(x)$ is differentiable at $a$, then $f$ is continuous at $a$.

#### Proof

Suppose $f$ is differentiable at $a$. Then $f'(a)$ exists and

$$f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$$

To show continuity at $a$, we must prove that $\lim_{x \to a} f(x) = f(a)$.

We can write:

$$\begin{align}
\lim_{x \to a} f(x) &= \lim_{x \to a} [f(x) - f(a) + f(a)] \\
&= \lim_{x \to a} [f(x) - f(a)] + \lim_{x \to a} f(a) \\
&= \lim_{x \to a} \left[ \frac{f(x) - f(a)}{x - a} \cdot (x - a) \right] + f(a) \\
&= \left( \lim_{x \to a} \frac{f(x) - f(a)}{x - a} \right) \cdot \left( \lim_{x \to a} (x - a) \right) + f(a) \\
&= f'(a) \cdot 0 + f(a) \\
&= f(a)
\end{align}$$

Therefore, $f$ is continuous at $a$. $\square$

### Important Note

**The converse is NOT true!** A function can be continuous at a point without being differentiable there.

## 7. When Derivatives Fail to Exist

A function may fail to be differentiable at a point for several reasons:

### Case 1: Discontinuity

If $f$ is not continuous at $a$, then $f$ is not differentiable at $a$.

**Example:** $f(x) = \begin{cases} x & x < 0 \\ x+1 & x \geq 0 \end{cases}$ is not continuous (hence not differentiable) at $x = 0$.

### Case 2: Corner or Cusp

If the left-hand and right-hand derivatives exist but are not equal, there is a "corner" or "cusp."

**Example:** $f(x) = |x|$ is continuous but not differentiable at $x = 0$ because:
- Left-hand derivative: $\lim_{h \to 0^-} \frac{|0+h| - |0|}{h} = \lim_{h \to 0^-} \frac{-h}{h} = -1$
- Right-hand derivative: $\lim_{h \to 0^+} \frac{|0+h| - |0|}{h} = \lim_{h \to 0^+} \frac{h}{h} = 1$

### Case 3: Vertical Tangent

If the tangent line is vertical, the derivative is undefined (slope would be infinite).

**Example:** $f(x) = \sqrt[3]{x}$ has a vertical tangent at $x = 0$ because:
$$f'(0) = \lim_{h \to 0} \frac{\sqrt[3]{h}}{h} = \lim_{h \to 0} \frac{1}{h^{2/3}} = \infty$$

### Case 4: Oscillatory Behavior

If the function oscillates too wildly near the point, the limit may not exist.

**Example:** $f(x) = \begin{cases} x\sin(1/x) & x \neq 0 \\ 0 & x = 0 \end{cases}$

The difference quotient at $x=0$ is:
$$\frac{f(h) - f(0)}{h} = \frac{h\sin(1/h)}{h} = \sin(1/h)$$

This oscillates between $-1$ and $1$ as $h \to 0$, so the limit doesn't exist.

## Summary of Key Definitions

| Concept | Definition |
|---------|-----------|
| **Difference Quotient** | $\frac{f(x) - f(a)}{x - a}$ or $\frac{f(a+h) - f(a)}{h}$ |
| **Derivative at $a$** | $f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}$ |
| **Alternative Form** | $f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}$ |
| **Instantaneous Velocity** | $v(a) = s'(a)$ where $s(t)$ is position |
| **Average Velocity** | $v_{\text{ave}} = \frac{s(t) - s(a)}{t - a}$ |
| **Differentiable at $a$** | $f'(a)$ exists |
| **Tangent Line Slope** | $m_{\text{tan}} = f'(a)$ |
