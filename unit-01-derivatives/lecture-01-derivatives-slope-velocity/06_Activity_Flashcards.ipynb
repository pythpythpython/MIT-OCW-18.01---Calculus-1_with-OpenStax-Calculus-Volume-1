### Card 1: What is the Derivative?

**Question**: Define the derivative of f(x) at x = a using limit notation.

**Answer**: The derivative of $f(x)$ at $x = a$ is:

$$f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}$$
